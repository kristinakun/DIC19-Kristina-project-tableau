```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv('merged_transaction.csv')
```

```python
data.head()
```

```python
to_delete = ['seller_id', 'order_approved_at', 'order_delivered_carrier_date', 'shipping_limit_date', 
             'product_name_lenght', 'product_description_lenght', 'product_photos_qty', 'product_weight_g', 
             'product_length_cm', 'product_height_cm', 'product_width_cm', 'product_category_name']

```

```python
for col in to_delete:
    if col in to_delete:
        data.drop(col, axis=1, inplace=True)
```

```python
data.head()
```

```python
#Adding country column for visualizations
data['country']='Brazil'
```

```python
#exporting prepared data
data.to_csv('clean_data.csv',index=False)
```
